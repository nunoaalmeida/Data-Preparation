### 2.2 Missing Values <a name="2.2"></a>

Missing data (or missing values) is the data value not filed for a variable in the observation of interest. The missing data problem is relatively common in almost all analyses and can significantly affect the conclusions drawn from the data. Accordingly Kang (2013), some studies have focused on handling the missing data, problems caused by missing data, and the methods to avoid or minimize. Changing missing Value in the same format (NaN) through the next function.

```python
missing_value_formats = ["n.a.","?","NA","n/a", "na", "--"]
scannia = pd.read_csv("aps_failure_set.csv", na_values = missing_value_formats)
```

```python
scannia
```

Using the function below, we can view the missing values and "0" values in the spreadsheet. Understanding that in our study case, positive numbers correlate with system failures and negative numbers are associated with out-of-system failures, a zero value would be a neutral result to identify the problem. Thus, we are going to intensify our studies on this issue to develop our EDA. This function create by ( Suhas_Pote, 2019 ) illustrates zero values, missing values, and their percentage.

```python
def missing_zero_values_table(df):
        zero_val = (df == 0.00).astype(int).sum(axis=0)
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mz_table = pd.concat([zero_val, mis_val, mis_val_percent], axis=1)
        mz_table = mz_table.rename(
        columns = {0 : 'Zero Values', 1 : 'Missing Values', 2 : '% of Total Values'})
        mz_table['Total Zero Missing Values'] = mz_table['Zero Values'] + mz_table['Missing Values']
        mz_table['% Total Zero Missing Values'] = 100 * mz_table['Total Zero Missing Values'] / len(df)
        mz_table['Data Type'] = df.dtypes
        mz_table = mz_table[
            mz_table.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns and " + str(df.shape[0]) + " Rows.\n"      
            "There are " + str(mz_table.shape[0]) +
              " columns that have missing values.")
#         mz_table.to_excel('D:/sampledata/missing_and_zero_values.xlsx', freeze_panes=(1,0), index = False)
        return mz_table

missing_zero_values_table(scannia)
```

### 3.3 Replace NAN ( MISSING VALUES ) using Interpolate mode <a name="3.3"></a>

"Interpolation is a technique in Python used to estimate unknown data points between two known data points. Interpolation is mostly used to impute missing values in the dataframe or series while preprocessing data."(Agrawal, 2021)

I can use several methods to substitute null values. We could use mean substitution, putting the value "0". However, the chosen one for this analysis was the linear method of Interpolate mode. This method ignores the index, treats missing values as evenly spaced, and finds the best point to fit the missing value after previous issues. If the missing value is in the first index, it will leave it as Nan. The reason for choosing this method was that, in our previous analysis, we could observe that the value "0" is present quite often in specific columns that, by replacing it, it could "influence" the result of the variable. Using the function df. interpolate from Pandas Librarie:


```python
Datascannia = scannia.interpolate(method='linear', inplace=False)
```

### 3.5 Outliers  (Q1 AND Q3) <a name="3.5"></a>

To start the analysis, we can say that with the previous steps, the presence of some Outliers has already been noticed. We filled in the missing values using the interpolating mode. Let's use the functions to identify the outliers in the Data Set. After performing the previous steps, we observed that the shape reduced ( 57143, 161)

According Hoppen (2018) an outlier is a value that escapes normality and can (and probably will) cause anomalies in the results obtained through algorithms and analytical systems.

In other words, they are unusual values that can distort the actual results and cause problems if not resolved.



```python
Q1 = scan_no_class.quantile(0.25) 
Q3 = scan_no_class.quantile(0.75) 
IQR = Q3 - Q1 
print(IQR)
```

```python
scan = scan_no_class[~((scan_no_class< (Q1-1.5 * IQR)) |(scan_no_class > (Q3 + 1.5 * IQR))).any(axis=1)] 
scan.shape
```

```python
pca = PCA().fit(scan)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of Features/Components")
plt.ylabel("Cumulative Explained Variance")
```

###  5.2 Retaining 99.5% variance in the data <a name="5.2"></a>

The graph presented to demonstrate cumulative Explained Variance above brings us the most significant number of Features/Components given in the Data Set (117 Components), and its variation remains constant at (1.00). ## 6. "Curse of Dimensionality" <a name="6"></a>


The curse of dimensionality, as the name says, refers to dimension. In terms of data and through the study that we had, we can conclude after all the steps previously done.


"The curse of dimensionality basically means that the error increases with the increase in the number of features. It refers to the fact that algorithms are harder to design in high dimensions and often have a running time exponential in the dimensions." (Choudhury, 2019)

Bring a reference so that we can elaborate a little more on the subject. It is possible to observe that the Scannia data brought sensitive data. We need to get them to values at more coherent scales so that we can analyze them better. The larger the data size, the greater the consequences and behavior of the algorithms. An expansion in the dimensionality in data results in data sparsification, and this exponential expansion is the first outcome that brings an introduction of a curse of dimensionality.
